```python
df['Department'] = df['Department'].fillna("Unknown")
df['Transaction ID'] = df['Transaction ID'].fillna('Missing')
```

```python
df['variance'] = df['Actual Amount'] - df['Budget Amount']
df['Variance %'] = df['variance'] / df['Budget Amount']
```

```python
df.head()
```

```python
total_budget = df['Budget Amount'].sum()
total_actual = df['Actual Amount'].sum()
total_variance = df['variance'].sum()
variance_percent = (total_variance / total_budget)

print("Total Budget:", total_budget)
print("Total Actual:", total_actual)
print("Total Variance:", total_variance)
print(f"Variance %: {variance_percent: .2%}")
```

```python
dept_analysis = df.groupby('Department').agg({
    'Budget Amount': 'sum',
    'Actual Amount': 'sum',
    'variance': 'sum'
}).reset_index()

print(dept_analysis)
```

```python
dept_analysis['Variance %'] = dept_analysis['variance'] / dept_analysis['Budget Amount']
dept_analysis = dept_analysis.sort_values(by='Variance %', ascending=False)
dept_analysis['Variance %'] = (dept_analysis['Variance %'] * 100).round(2)


dept_analysis
```

```python
top_dept = dept_analysis.iloc[0]
print("Highest Overspending Department:")
print(top_dept)
```

```python
dept_analysis = df.groupby('Department').agg({
    'Budget Amount': 'sum',
    'Actual Amount': 'sum',
    'variance': 'sum'
}).reset_index()

dept_analysis['variance %'] = dept_analysis['variance'] / dept_analysis['Budget Amount']

dept_analysis = dept_analysis.sort_values(by='variance %', ascending=False)

dept_analysis['variance %'] = (dept_analysis['variance %'] * 100).round(2)

dept_analysis
```

```python
category_analysis = df.groupby('Category').agg({
    'Budget Amount': 'sum',
    'Actual Amount': 'sum',
    'variance': 'sum'
}).reset_index()

category_analysis
```

```python
category_analysis['Variance %'] = category_analysis['variance'] / category_analysis['Budget Amount']
category_analysis = category_analysis.sort_values(by='Variance %', ascending=False)
category_analysis['Variance %'] = (category_analysis['Variance %'] * 100).round(2)

category_analysis
```

```python
top_category = category_analysis.iloc[0]
print("Top Cost Driver:")
print(top_category)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
```

```python
df[['Date', 'Year']].head()
```

```python
year_analysis = df.groupby('Year').agg({
    'Budget Amount': 'sum',
    'Actual Amount': 'sum',
    'variance': 'sum'
}).reset_index()
```

```python
year_analysis['Variance %'] = year_analysis['variance'] / year_analysis['Budget Amount']
year_analysis = year_analysis.sort_values(by='Year')
year_analysis = year_analysis.sort_values(by='Year')
year_analysis['Variance %'] = (year_analysis['Variance %'] * 100).round(2)
year_analysis
```

```python
df['Month'] = df['Date'].dt.month
df['Month Name'] = df['Date'].dt.strftime('%b')
```

```python
monthly_analysis = df.groupby(['Year', 'Month', 'Month Name']).agg({
    'Budget Amount': 'sum',
    'Actual Amount': 'sum',
    'variance': 'sum'
}).reset_index()
```

```python
monthly_analysis['Variance %'] = monthly_analysis['variance'] / monthly_analysis['Budget Amount']
monthly_analysis = monthly_analysis.sort_values(by=['Year', 'Month'])
monthly_analysis = monthly_analysis.sort_values(by=['Year', 'Month'])
monthly_analysis['Variance %'] = (monthly_analysis['Variance %'] * 100).round(2)
monthly_analysis.head(15)
```

```python
monthly_2022 = monthly_analysis[monthly_analysis['Year'] == 2022]
monthly_2022
```
